```python
from mua.train import regression
from mua.plots import unitary
from mua.plots import fit
import numpy as np
import os
from pathlib import Path

folder = Path('./examples/fits/')

os.makedirs(folder, exist_ok = True)
```

```python
depth = 5
num_epochs = 150
lr = 0.1

def tanh(x):
    return np.tanh(5 * x)

params, params_history, loss_history = regression.train(tanh, depth, num_epochs, lr, folder)
```

```python
fit.show(tanh, params, save = str(folder / 'fit1.pdf'))
```

```python
depth = 10
num_epochs = 150
lr = 0.1

def upsin(x):
    return (x + np.sin(3*np.pi*x))/2

params, params_history, loss_history = regression.train(upsin, depth, num_epochs, lr, folder)
```

```python
fit.show(upsin, params, save = str(folder / 'fit2.pdf'))
```

```python
depth = 50
num_epochs = 1000
lr = 0.001

def squarewave(x):
    return np.sign(np.cos(2*np.pi*x))

params, params_history, loss_history = regression.train(squarewave, depth, num_epochs, lr, folder)
```

```python
fit.show(squarewave, params, save = str(folder / 'fit3.pdf'))
```

```python
depth = 50
num_epochs = 1000
lr = 0.002

def veryhard(x):
    return np.sign(np.cos(2*np.pi*x))*np.abs(x)

params, params_history, loss_history = regression.train(veryhard, depth, num_epochs, lr, folder)
```

```python
fit.show(veryhard, params, save = str(folder / 'fit4.pdf'))
```
